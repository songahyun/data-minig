# 지도 시각화를 위해 필요한 컬럼만 남겨서 새로운 데이터프레임 생성

```python
# 남기고 싶은 구 리스트 정의
seoul_regions = [
    "강서구", "강남구", "동작구", "영등포구", "은평구", "도봉구", "강북구", "관악구",
    "노원구", "금천구", "중랑구", "광진구", "서대문구", "송파구", "성동구",
    "성북구", "마포구", "양천구", "강동구", "용산구", "중구", "종로구",
    "동대문구", "서초구", "구로구"
]

# 출발지구와 목적지구가 모두 valid_zones에 포함된 행만 남기기
df = df[(df['출발지구'].isin(seoul_regions)) & (df['목적지구'].isin(seoul_regions))]
```

```python
# 출발지구별 데이터 카운트
departure_counts = df['출발지구'].value_counts().reset_index()
departure_counts.columns = ['출발지구', '카운트']

# 목적지구별 데이터 카운트
destination_counts = df['목적지구'].value_counts().reset_index()
destination_counts.columns = ['목적지구', '카운트']

# 결과 확인
print("출발지구별 카운트")
print(departure_counts)
print("\n목적지구별 카운트")
print(destination_counts)
```

# 출발지구 지도 시각화

## 출발지구 표시X

```python
# 서울시 구 단위 GeoJSON 파일 불러오기
geo_data = gpd.read_file('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')

# GeoDataFrame 병합: '출발지구' 컬럼을 기준으로 병합
geo_data = geo_data.rename(columns={'name': '출발지구'})
merged = geo_data.merge(departure_counts, on='출발지구', how='left')

# Folium 지도 만들기 (중심 좌표는 서울의 중앙으로 설정)
m = folium.Map(
    location=[37.5665, 126.9780],  # 서울 중앙
    zoom_start=11,  # 초기 줌 레벨
    scrollWheelZoom=False,  # 마우스 휠로 확대/축소 비활성화
    dragging=False,  # 드래그로 지도 이동 비활성화
    zoomControl=False  # 줌 컨트롤 비활성화
)

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=merged.to_json(),
    data=merged,
    columns=['출발지구', '카운트'],
    key_on='feature.properties.출발지구',  # GeoJSON에서 '출발지구' 매핑
    fill_color='YlOrRd',
    fill_opacity=1,  # 투명도 없애기
    line_opacity=0.5,  # 구경계선의 투명도 조정
    legend_name='출발지구별 카운트',
    highlight=True,  # 마우스를 올리면 강조
    nan_fill_color='white',  # 데이터가 없는 지역은 하얀색
).add_to(m)

# 지도 저장
m.save('seoul_departure_map_without_names.html')

m  # 지도 출력
```

## 출발지역 표시

```python
# 서울시 구 단위 GeoJSON 파일 불러오기
geo_data = gpd.read_file('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')

# GeoDataFrame 병합: '출발지구' 컬럼을 기준으로 병합
geo_data = geo_data.rename(columns={'name': '출발지구'})
merged = geo_data.merge(departure_counts, on='출발지구', how='left')

# Folium 지도 만들기 (중심 좌표는 서울의 중앙으로 설정)
m = folium.Map(
    location=[37.5665, 126.9780],  # 서울 중앙
    zoom_start=11,  # 초기 줌 레벨
    scrollWheelZoom=False,  # 마우스 휠로 확대/축소 비활성화
    dragging=False,  # 드래그로 지도 이동 비활성화
    zoomControl=False  # 줌 컨트롤 비활성화
)

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=merged.to_json(),
    data=merged,
    columns=['출발지구', '카운트'],
    key_on='feature.properties.출발지구',  # GeoJSON에서 '출발지구' 매핑
    fill_color='YlOrRd',
    fill_opacity=1,  # 투명도 없애기
    line_opacity=0.5,  # 구경계선의 투명도 조정
    legend_name='출발지구별 카운트',
    highlight=True,  # 마우스를 올리면 강조
    nan_fill_color='white',  # 데이터가 없는 지역은 하얀색
).add_to(m)

# 각 구에 대한 이름을 지도에 바로 표시 (텍스트 추가)
for _, row in merged.iterrows():
    # 각 구의 중심 좌표 계산
    lat = row['geometry'].centroid.y
    lon = row['geometry'].centroid.x
    
    # 텍스트로 마커 대신 DivIcon 사용
    folium.Marker(
        location=[lat, lon],
        icon=folium.DivIcon(html=f'<div style="font-size: 12px; color: {color};">{row["출발지구"]}</div>'),
        popup=row['출발지구'],  # 팝업에 구 이름 표시
        tooltip=row['출발지구'],  # 마우스를 올리면 구 이름 툴팁 표시
    ).add_to(m)

# 지도 저장
m.save('seoul_departure_map_with_names.html')

m  # 지도 출력
```
